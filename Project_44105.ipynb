## <span style='color:brown'><span style='font-family:Georgia'> ***20. Find the data of maximum and minimum powerplant based on the given country.***

```python
# Here is the code for the maximum and the minimum value for the specific data.
def min_max_values(col):
  
    top = UK[col].idxmax()
    top_obs = pd.DataFrame(UK.loc[top])
    
    bottom = UK[col].idxmin()
    bot_obs = pd.DataFrame(UK.loc[bottom])
    
    min_max_obs = pd.concat([top_obs, bot_obs], axis = 1)
    return min_max_obs
```

```python
# Here we can see that the powerplant with id 156 has maximum power it is 1728000kw and id 717 has minimum power of 2kw.
min_max_values('Rated Power (kW)')
# We can show different data also for the maximum and minimum values.
```
